```python
import re

# --- Intent Recognition and Query Handling ---

def find_customer_by_id_and_name(custid, name):
    for customer in bank_customers:
        if (customer['customer_id'].lower() == custid.lower() and
            customer['name'].lower() == name.lower()):
            return customer
    return None

def get_account_balances(customer):
    lines = []
    for acc in customer['accounts']:
        lines.append(f"{acc['account_type']} ({acc['account_number']}): {acc['currency']} {acc['balance']}")
    return "\n".join(lines)

def get_branch_info(customer):
    branches = set(acc['branch'] for acc in customer['accounts'])
    return f"Your account(s) are at: {', '.join(branches)}"

def get_account_types(customer):
    types = set(acc['account_type'] for acc in customer['accounts'])
    return f"You have the following account types: {', '.join(types)}"

def get_account_numbers(customer):
    nums = [acc['account_number'] for acc in customer['accounts']]
    return f"Your account number(s): {', '.join(nums)}"

def handle_query(customer, user_input):
    text = user_input.lower()
    # 1. Check FAQ first
    faq_answer = handle_faq(text)
    if "Sorry, I can help you" not in faq_answer:
        return faq_answer

    # 2. Customer-specific queries
    if "balance" in text:
        return get_account_balances(customer)
    elif "branch" in text:
        return get_branch_info(customer)
    elif "address" in text:
        return f"Your registered address is: {customer['address']}"
    elif "email" in text:
        return f"Your registered email is: {customer['email']}"
    elif "phone" in text or "contact" in text or "mobile" in text:
        return f"Your registered phone number is: {customer['phone']}"
    elif "account type" in text or "types of account" in text:
        return get_account_types(customer)
    elif "account number" in text:
        return get_account_numbers(customer)
    elif "ifsc" in text:
        ifscs = set(acc['ifsc'] for acc in customer['accounts'])
        return f"Your account IFSC code(s): {', '.join(ifscs)}"
    elif "credit card" in text:
        cards = [acc for acc in customer['accounts'] if acc['account_type'].lower() == "credit card"]
        if cards:
            card = cards[0]
            return f"Your credit card ({card['account_number']}) outstanding balance is {card['currency']} {abs(card['balance'])}"
        else:
            return "You do not have a credit card account with us."
    elif "fixed deposit" in text:
        fds = [acc for acc in customer['accounts'] if acc['account_type'].lower() == "fixed deposit"]
        if fds:
            fd = fds[0]
            return f"Your fixed deposit ({fd['account_number']}) balance is {fd['currency']} {fd['balance']}"
        else:
            return "You do not have a fixed deposit account with us."
    elif "recurring deposit" in text:
        rds = [acc for acc in customer['accounts'] if acc['account_type'].lower() == "recurring deposit"]
        if rds:
            rd = rds[0]
            return f"Your recurring deposit ({rd['account_number']}) balance is {rd['currency']} {rd['balance']}"
        else:
            return "You do not have a recurring deposit account with us."
    else:
        return "Sorry, I didn't understand your query. Please ask about your account or a general banking question."

def handle_faq(text):
    for question, answer in faq_data.items():
        if question in text:
            return answer
    return "Sorry, I can help you with balance, branch, account details, and common FAQs. Please ask accordingly."


# --- Chatbot Conversation Loop ---

def main():
    print("Welcome to HDFC Bank Chatbot! Please authenticate yourself.")
    customer = None
    while not customer:
        custid = input("Enter your Customer ID: ").strip()
        name = input("Enter your Name: ").strip()
        customer = find_customer_by_id_and_name(custid, name)
        if not customer:
            print("Chatbot: Sorry, authentication failed. Please check your Customer ID and Name and try again.")
        else:
            print(f"Chatbot: Welcome, {customer['name']}! How can I assist you today?")
            print(f"For exiting chat section input ['exit', 'quit', ':q']")
    while True:
        user_input = input("You: ")
        if user_input.lower() in ["exit", "quit", ":q"]:
            print("Chatbot: Thank you for using HDFC Bank Chatbot. Goodbye!")
            break
        response = handle_query(customer, user_input)
        print("Chatbot:", response)

if __name__ == "__main__":
    main()
```
